```python
sales_data = pd.read_csv('/Users/master/Downloads/Monthly Ice Cream Shop Sales by Flavor.csv')

melted_data = pd.melt(sales_data, id_vars=['Month', 'Total Units', 'Total Profit'],
                      var_name='Ice Cream Flavor', value_name='Units Sold')

sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))


sns.lineplot(x='Month', y='Units Sold', hue='Ice Cream Flavor', data=melted_data, marker='o', palette='Set1', errorbar=None, alpha=0.5)

plt.title('Units Sold by Ice Cream Flavor (Monthly)')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend(title='Ice Cream Flavor', loc='upper right')
plt.show()

## Chocolate and Lime have the exact same monthly sales numbers - tried to play around with transparency but it's still not super visible. ##
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Total Profit', data=sales_data, marker='o', color='blue')
plt.title('Total Profit Across Months')
plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Units Sold', hue='Ice Cream Flavor', data=melted_data, palette='Set1')
plt.title('Total Units Sold Across Months by Ice Cream Flavor')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.legend(title='Ice Cream Flavor')
plt.grid(axis='y')
plt.show()
```
